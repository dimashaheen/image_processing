```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow #to display images
```

```python
img = cv2.imread("lena.jpg")
cv2_imshow(img)
```

```python
col = img.shape[0]
row = img.shape[1]
col=col//2
row=row//2
print(col,row)
```

```python
center_coordinates = (col, row)
radius = 25
color = (255, 0, 0)
thickness = 2
image = cv2.circle(img, center_coordinates, radius, color, thickness)
cv2_imshow(image)
```

```python
image2=cv2.line(img,(0,0),(256,256),(0,255,0),5)
cv2_imshow(image2)
```

```python
image3=cv2.rectangle(img,(256,256),(512,512),(0,0,255),2)
cv2_imshow(image3)
```

```python
from matplotlib import pyplot as plt
img = cv2.imread('mandril_c.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
