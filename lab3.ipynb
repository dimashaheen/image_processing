```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/lena.jpg')
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```python
(h,s,v) = cv2.split(hsv)
cv2_imshow(h)
cv2_imshow(s)
cv2_imshow(v)
```

```python
histh = cv2.calcHist([h],[0],None,[256],[0,256])
hists = cv2.calcHist([s],[0],None,[256],[0,256])
histv = cv2.calcHist([v],[0],None,[256],[0,256])

plt.plot(histh, color='blue')
plt.plot(hists, color='green')
plt.plot(histv, color='red')
plt.show()
```
