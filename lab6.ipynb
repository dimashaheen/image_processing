***Edge Detection***

***Sobel (X-Direct Kernel / Y-Direct Kernel)***
Sobel(img , ddepth , dx , dy , ksize )

Steps : 
1.   Read Image
2.   Gray Conversion
3.   Noise Filter using "Gaussian"
4.   Edge Detector


```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.jpg')
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
image = cv2.GaussianBlur(gray, (3,3), 1)
cv2_imshow(image)
```

Detect horizontal lines

```python
sobel_Y = cv2.Sobel(image,cv2.CV_64F,0,1 , ksize= 3)
cv2_imshow(sobel_Y)
```

Detect Vertical lines

```python
sobel_X = cv2.Sobel(image,cv2.CV_64F,1,0 , ksize= 3)
cv2_imshow(sobel_X)
```

```python
sobel_XY = cv2.Sobel(image,cv2.CV_64F,1,1  , ksize= 3)
cv2_imshow(sobel_XY)
```

Canny Steps:
Best of edge detectors

1.   Noise Reduction using "Gaussian"
2.   Intensity Gradient XY sobel
3.   Non maximum suppression (compare gradient magnitudes)
4.   Thresholding (Upper , Lower values)  
     grad mag > upper --> strong edge
     grad mag < lower --> not an edge
     between range --> weak edge

Canny(image , upper , lower)

```python
canny = cv2.Canny(gray , 100 , 200)
cv2_imshow(canny)
```
