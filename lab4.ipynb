```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("backlighting.jpg")
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #to reduce dimenstions 
```

```python
#apply threshold value
ret,image2=cv2.threshold(gray,200,255,cv2.THRESH_BINARY) #better to use 127 instead of 200
cv2_imshow(image2)
```

```python
ret,image3 = cv2.threshold(gray, 0, 255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
cv2_imshow(image3)
```

```python
equ = cv2.equalizeHist(gray)
res = np.hstack((gray,equ)) #stacking images side-by-side
cv2_imshow(res)
```
