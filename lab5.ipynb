```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('backlighting.jpg')
#cv2_imshow(img)
```

```python
def gammaCorrection(img, gamma):
    invGamma = 1 / gamma
##create table of 255 to loop over every pixelin image and place new value in table 
    table = [((i / 255) ** invGamma) * 255 for i in range(256)]
    table = np.array(table, np.uint8)
 
    return cv2.LUT(img, table)
 
##as gamma increase brightness increases
gammaImg = gammaCorrection(img, 2.4)
cv2_imshow(gammaImg)
```

```python
##Blur use kernel  (almost averaging)
##Meadian kernek has to be odd cv2.MedianBlur
##guassian used if we have guassian noise
img2 = cv2.imread('moon_shot_noise.jpg')
cv2_imshow(img2)
img2.shape
```

```python
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY) #to reduce dimenstions
image = cv2.GaussianBlur(gray, (5,5), 1)
cv2_imshow(image)
```

```python
blurImg = cv2.blur(gray,(5,5)) 
cv2_imshow(blurImg)
```

```python
img_median=cv2.medianBlur(gray,5)
cv2_imshow(img_median)
```
